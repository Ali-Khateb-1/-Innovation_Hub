```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# ÿßŸÑŸÖÿ≥ÿßÿ± ÿßŸÑÿ£ÿ≥ÿßÿ≥Ÿä ŸÑŸÖÿ¨ŸÑÿØ ÿßŸÑÿµŸàÿ±
test_images_path = "/mnt/c/Users/ali/Desktop/intro-to-ml-tensorflow/projects/p2_image_classifier/test_images"

# ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿπÿ±ÿ∂ ÿßŸÑÿµŸàÿ±
image_files = ['cautleya_spicata.jpg', 'hard-leaved_pocket_orchid.jpg', 'orange_dahlia.jpg', 'wild_pansy.jpg']

fig, axes = plt.subplots(1, 4, figsize=(15, 5))

for i, image_file in enumerate(image_files):
    img = mpimg.imread(os.path.join(test_images_path, image_file))
    axes[i].imshow(img)
    axes[i].axis('off')
    axes[i].set_title(image_file)

plt.show()
```

```python
import os

# ÿ™ÿ≠ÿØŸäÿØ ŸÖÿ≥ÿßÿ± ÿßŸÑŸÖÿ¨ŸÑÿØ
folder_path = "C:/Users/ali/Desktop/Innovation_Hub_New/Full_Stack_Framework"

# ÿßÿ≥ÿ™ÿπÿ±ÿßÿ∂ ÿßŸÑŸÖŸÑŸÅÿßÿ™ ŸàÿßŸÑŸÖÿ¨ŸÑÿØÿßÿ™ ÿØÿßÿÆŸÑ ÿßŸÑŸáŸäŸÉŸÑŸäÿ©
def list_directory_structure(path, indent=0):
    try:
        for item in os.listdir(path):
            item_path = os.path.join(path, item)
            if os.path.isdir(item_path):
                print("  " * indent + f"[Folder] {item}")
                list_directory_structure(item_path, indent + 1)
            else:
                print("  " * indent + f"- {item}")
    except Exception as e:
        print(f"Error accessing {path}: {e}")

# ÿ™ŸÜŸÅŸäÿ∞ ÿßŸÑÿ™ÿ≠ŸÑŸäŸÑ
list_directory_structure(folder_path)
```
