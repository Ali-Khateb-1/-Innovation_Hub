## 🔧 **تنفيذ عمليات تكامل البيانات**
```python
import pandas as pd

# تحميل البيانات من ملف CSV
data_path = "/mnt/c/Users/ali/Desktop/Innovation_Hub_New/Data/data_source.csv"
df = pd.read_csv(data_path)

# تنظيف البيانات وإزالة القيم الفارغة
df_cleaned = df.dropna()

# تحويل البيانات إلى صيغ قابلة للاستخدام داخل نماذج الذكاء الاصطناعي
df_normalized = (df_cleaned - df_cleaned.mean()) / df_cleaned.std()

print("✅ تم تجهيز البيانات بنجاح!")
```
